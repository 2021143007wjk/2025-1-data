```python
with open('전국휴양림표준데이터.json','r') as files:
  temp = json.load(files)
temp
```

```python
n_data = []
for d in temp['records']:
  n_data.append(d)
n_data
```

```python
pd_data = pd.DataFrame(n_data)
pd_data
```

```python
outputFile = '자연휴양.xlsx'
pd_data.to_excel(outputFile)
```

```python
pd_data.head()
```

```python
import folium
rest_map = folium.Map(location=[38.14199686, 128.0710755],
                      zoom_start=11,
                      zoom_control=False,
                      control_scale=True)
rest_map
```

```python
for n, row in pd_data.iterrows():
  mk = folium.Marker(location=[row['위도'], row['경도']], popup=folium.Popup(row['휴양림명'], max_width=200))
  mk.add_to(rest_map)
rest_map
```

```python
from folium.plugins import MarkerCluster

rest_map2 = folium.Map(location=[38.14199686, 128.0710755],
                      zoom_start=10,
                      zoom_control=False,
                      control_scale=True)

mc = MarkerCluster().add_to(rest_map2)

for n, row in pd_data.iterrows():
  folium.Marker(location=[row['위도'], row['경도']],
                popup=folium.Popup(row['휴양림명'], max_width=200)).add_to(mc)

rest_map2
```
